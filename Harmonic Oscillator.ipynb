# Wigner Function of Harmonic Oscilltor

Hamiltonian of Harmonic Oscillator:

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import qutip
# Hamiltonian
N = 20       # num. of levels
omega = 1    # frequency
H = omega*qutip.destroy(N)*qutip.create(N)
```

Wigner function for the Hamiltonian's eigen states:

```python
# find its eigen energies and states
eig_energy = H.eigenenergies()
eig_states = H.eigenstates()
# Plot wigner function for lowest 3 states
x = qutip.destroy(N)+qutip.create(N)
xvec = np.linspace(-5,5,200)
fig0, axes0 = plt.subplots(1, 3, figsize=(12,3))
fig1, axes1 = plt.subplots(1, 3, figsize=(12,3), subplot_kw={'projection':'3d'})
for i in range(3):
    W = qutip.wigner(eig_states[1][i], xvec, xvec)
    cont = axes0[i].contourf(xvec, xvec, W, 100)
    lbl = axes0[i].set_title("state " + str(i))
    cont = axes1[i].contourf(xvec, xvec, W, 100)
    lbl = axes1[i].set_title("state " + str(i))
```

Wigner function for the coherent states:

```python
n_states = 5
xvec = np.linspace(-10,10,200)
fig0, axes0 = plt.subplots(1, n_states, figsize=(16,3))
fig1, axes1 = plt.subplots(1, n_states, figsize=(16,3), subplot_kw={'projection':'3d'})
for i in range(n_states):
    state = qutip.coherent(N, i)
    W = qutip.wigner(state, xvec, xvec)
    cont = axes0[i].contourf(xvec, xvec, W, 100)
    lbl = axes0[i].set_title("state " + str(i))
    cont = axes1[i].contourf(xvec, xvec, W, 100)
```
