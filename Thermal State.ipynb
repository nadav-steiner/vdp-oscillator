# Thermal State Simulation

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import qutip
```

```python
N = 20
n_states = 5
fig0, axes0 = plt.subplots(1, n_states,figsize=(12,3))
fig1, axes1 = plt.subplots(1, n_states,figsize=(12,3), subplot_kw={'projection':'3d'})
for i in range(n_states):
    state = qutip.thermal_dm(N, i)
    qutip.plot_wigner(state, fig=fig0, ax=axes0[i])
    qutip.plot_wigner(state, fig=fig1, ax=axes1[i], projection='3d')
```
